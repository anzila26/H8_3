Preprocessing

```python
#menghapus kolom unnamed:0 dan kolom EASE-MENT
df.drop(["Unnamed: 0", "EASE-MENT"], axis=1, inplace=True)
```

```python
#mengubah tipe data kolom SALE DATE menjadi date time
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'], errors='coerce')
```

```python
#memgubah tipe data menjadi category
categorical = ["BOROUGH","NEIGHBORHOOD",'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT','ZIP CODE', 
               'YEAR BUILT', 'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE']

for col in categorical: 
    df[col] = df[col].astype("category")
```

```python
#mengubah tipe data menjadi numeric
numeric = ['LAND SQUARE FEET','GROSS SQUARE FEET', 'SALE PRICE']
for col in numeric : 
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

Measure of Central Tendency

```python
#menghitung nilai mean pada kolom TOTAL UNITS pada setiap data dikolom BOROUGH
mean_totunits = df[['BOROUGH', 'TOTAL UNITS']].groupby('BOROUGH').mean().sort_values(by='TOTAL UNITS', ascending=True)
mean_totunits
```

Berdasarkan hasil mean yang ditampilkan, dapat diketahui bahwa rata-rata tertingi total units yaitu terdapat pada kota The Bronx yaitu sebanyak 22.134715

```python
#menghitung median pada kolom TOTAL UNITES pada setiap data dikolom BOROUGH
median_totunits = df[['BOROUGH','TOTAL UNITS']].groupby('BOROUGH').median().sort_values(by='TOTAL UNITS', ascending=True)
median_totunits
```

Berdasarkan hasil diatas, dapat disimpulkan bahwa median tertinggi yaitu terdapat pada kota The Bronx yitu dengan nilai 6, sedangkan yang lainnya sama dengan nilai 1

```python
#menghitung nilai modus TOTAL UNITS pada setiap kolom BOROUGH
modus_totunits = df[['BOROUGH','TOTAL UNITS']].groupby(['BOROUGH']).apply(pd.DataFrame.mode).reset_index(drop=True)
modus_totunits
```

Berdasarkan hasil diatas, dapat disimpulkan bahwa sebagian besar pada masing-masing kota memiliki 1 unit.

```python
#menghitung nilai range TOTAL UNITS pada setiap data di kolom BROUGH
range_totunits = df.groupby('BOROUGH').apply(lambda x: x['TOTAL UNITS'].max() - x['TOTAL UNITS'].min())
range_totunits
```

berdasarkan data diatas, dapat disimpulkan bahwa kota queens memiliki range paling tinggi dintara kota lainnya yaitu berada di angka 2261. dan range paling bawah ditempati oleh kota staten island di angka 143

Probability Distribution

```python
#membuat probability distribution Staten Island di kolom SALE PRICE
df1 = df.groupby("BOROUGH")
br = df1.get_group('Staten Island')

ax = sns.distplot(br['SALE PRICE'],
                  kde=True,
                  bins=20,
                  color='green',
                  hist_kws={"linewidth": 15,'alpha':0.45})

ax.set(xlabel='Price', 
       ylabel='Probability of Occurence', 
       title='Distribution of Sale Price in Staten Island')
```

Berdasarkan hasil plot distribusi di atas, dapat disimpulkan bahwa SALE PRICE State Island lebih condong ke kanan. Dengan kata lain, nilai-nilai observasi berfrekuensi rendah lebih banyak berada di sebelah kanan dari rata-rata.

Confidence Intervals

```python
#menampilkan mean, standard deviation, dan population size dari SALE PRICE
ci=df.groupby("BOROUGH").agg({"SALE PRICE": [np.mean, np.std, np.size]})
ci
```

```python
#Mengekstrak parameter yang diperlukan
mean_sp = ci.iloc[0,0]
sd = ci.iloc [0,1]
n = ci.iloc [0,2]         
z = 1.96    
```

```python
#Menghitung standard error menggunakan formula untuk standard error dari mean
se = sd / np.sqrt(n)
```

```python
#Membangun CI
lcb = mean_sp - z * se
ucb = mean_sp + z * se
lcb, ucb
```

Berdasarkan data diatas, mengartikan bahwa rata-rata sebenarnya dari SALE PRICE di The Bronx akan jatuh antara 13549150.798785087 dan 24865663.704841856

Hypothesis Testing

```python
#Shapiro-Wilk Test
from scipy.stats import pearsonr
df1 = df.groupby("BOROUGH")
mh = df1.get_group('The Bronx')
sp = mh['SALE PRICE']

from scipy.stats import shapiro
stat, p = shapiro(sp)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

```python
#D'Agostino's K2 Test
from scipy.stats import pearsonr
df1 = df.groupby("BOROUGH")
br = df1.get_group('Brooklyn')
sp1 = br['SALE PRICE']

from scipy.stats import normaltest
stat, p = normaltest(sp1)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

```python
#Anderson-Darling Test
from scipy.stats import pearsonr
df1 = df.groupby("BOROUGH")
brok = df1.get_group('Manhattan')
sp2 = brok['SALE PRICE']

from scipy.stats import anderson
results = anderson(sp2)
print('stat=%.3f' % (results.statistic))
for i in range(len(results.critical_values)):
      sl, cv = results.significance_level[i], results.critical_values[i]
      if results.statistic < cv:
          print('Probably Gaussian at the %.1f%% level' % (sl))
      else:
          print('Probably not Gaussian at the %.1f%% level' % (sl))
```

Berdasarkan data diatas, dapat disimpulkan dari p-value yang didapatkan <0,05 sehingga H0 ditolak yang mengartikan sempel tidak berdistribusi gaussian.
